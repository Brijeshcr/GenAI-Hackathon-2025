```python
import json

import boto3
from botocore.exceptions import ClientError
```

```python
session = boto3.Session(profile_name="bedrock-hackathon")
bedrock_runtime = session.client("bedrock-runtime")
boto3_bedrock = session.client('bedrock')
```

```python
# List all available foundation models
[models['modelId'] for models in boto3_bedrock.list_foundation_models()['modelSummaries']]
```

```python
# Choose a model from the list of models
model_id = "us.anthropic.claude-3-7-sonnet-20250219-v1:0"
```

```python
# Create a prompt
prompt = "Write a random story"
```

```python
# Prepare the request payload
native_request = {
            "anthropic_version": "bedrock-2023-05-31",
            "max_tokens": 1000,
            "system": "",
            "messages": [{"role": "user", "content": prompt}]
        }  
request = json.dumps(native_request)
```

```python
# Invoke the model with the request.
try:
    response = bedrock_runtime.invoke_model(modelId=model_id, body=request)

except (ClientError, Exception) as e:
    print(f"ERROR: Can't invoke '{model_id}'. Reason: {e}")
    exit(1)
```

```python
# Parse the model response
model_response = json.loads(response["body"].read())
```

```python
# Print the response text
response_text = model_response["content"][0]["text"]
print(response_text)
```
